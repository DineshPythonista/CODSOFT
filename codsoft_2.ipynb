```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("IMDb Movies India.csv", encoding='latin-1')
```

```python
df.head()
```

```python
df.columns
```

```python
col = [ 'Genre', 'Duration',  'Votes', 'Director',
       'Actor 1', 'Actor 2', 'Actor 3','Rating']
df = df[col]
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['Duration'] = pd.to_numeric(df['Duration'].str.replace(' min', ''))
```

```python
df["Votes"] = pd.to_numeric(df["Votes"].str.replace('[^\d.]', '', regex=True))
df["Votes"].fillna(df["Votes"].mean())
```

```python
df["Director"] = df["Director"].astype("category")
df["Director"] = df["Director"].cat.codes
```

```python
df["Genre"] = df["Genre"].astype("category")
df["Genre"] = df["Genre"].cat.codes
df["Actor 1"] = df["Actor 1"].astype("category")
df["Actor 1"] = df["Actor 1"].cat.codes
df["Actor 2"] = df["Actor 2"].astype("category")
df["Actor 2"] = df["Actor 2"].cat.codes
df["Actor 3"] = df["Actor 3"].astype("category")
df["Actor 3"] = df["Actor 3"].cat.codes
```

```python
df["Votes"] = df["Votes"].fillna(df["Votes"].mean())
df["Rating"] = df["Rating"].fillna(df["Rating"].mean())
df["Duration"] = df["Duration"].fillna(df["Duration"].mean())
```

```python
df.head()
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state = 42, test_size = 0.3)
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_pred)
print("Mean absolute error", mae)
```

```python

data = np.column_stack((x_test, y_pred))
df1 = pd.DataFrame(data, columns=df.columns)
```
